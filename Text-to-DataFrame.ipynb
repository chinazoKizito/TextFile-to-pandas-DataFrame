### DATA LOADING AND CLEANING 

```python
fhand = open("get-me-some.txt")
order = []
name = []
job = []
for line in (lin.rstrip() for lin in fhand):
    a = line.split(',')
    order.append(a[0][12:])
    b = a[1].split()
    name.append(b[1])
    job.append(b[3]) 
```

### PRESENTATION OF THE MOST IMPORTANT ATTRIBUTES IN  THE SENTENCES IN A PANDAS DATAFRAME

```python
import pandas as pd
df = pd.DataFrame({"Name":name, "Workplace": job, "Order":order})
df
```

#### CREATING A BAR CHAT OF ALL THE WORDS IN THE TEXT FILE JUST FOR FUN

#### Counting the words

```python
words_count ={}
with open('get-me-some.txt') as f:
    for line in (lin.rstrip() for lin in f):
        a = line.split()
        for word in a:
            if word in words_count:
                words_count[word] = words_count[word]+1
            else:
                words_count[word] =1
```

#### Sorting the words

```python
word_sort = []
for key, values in words_count.items():
    word_sort.append((values, key))
word_sort.sort(reverse=True)
```

#### Putting the Sorted words in a Pandas Dataframe

```python
word_df  = pd.DataFrame(word_sort, columns=["Number", "Word"])
word_df.head(10)
```

#### Visualizing the first 20 highest occuring words in descending order

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.style as style
style.available
style.use("seaborn-poster")
style.use("ggplot")



fig, axes = plt.subplots()
axes.barh(word_df["Word"][:20], word_df["Number"][:20])
plt.gca().invert_yaxis()
```

Clearly `['some', "me", "get", "at"]` are clearly the highest occuring words in the text file at number 45 each. as they were the words used in all the sentences. While `[Marcus, Ekene and Ben]` appeared twice in the names counted. Other words came only once.
